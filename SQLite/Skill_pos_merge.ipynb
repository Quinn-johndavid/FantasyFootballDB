## Step 1 clean

These csv's have two targets, the sqlite db that this script creates, and to clean the data for my postgres local server.) No more manual cleaning of the data if this gets elaborated upon!

```python
qbdataroute = "../Resources/FantasyPros_Fantasy_Football_Statistics_QB.csv"
qbdataDF = pd.read_csv(qbdataroute)
qbdataDF
```

```python
qbdataDF['QBID'] = "QB"
cleanqbdataDF = qbdataDF.rename(columns={'QBID':'Position'})
cleanqbdataDF['PassingYDS'] = cleanqbdataDF['PassingYDS'].str.replace(',', '').astype(float)
cleanqbdataDF['RushingYDS'] = cleanqbdataDF['RushingYDS'].str.replace(',', '').astype(float)
cleanqbdataDF['ROST'] = cleanqbdataDF['ROST'].str.rstrip('%').astype(float)
cleanqbdataDF.dropna(inplace=True)
cleanqbdataDF
```

```python
wrdataroute = "../Resources/FantasyPros_Fantasy_Football_Statistics_WR.csv"
wrdataDF = pd.read_csv(wrdataroute)
wrdataDF
```

```python
wrdataDF['WRID'] = "WR"
cleanwrdataDF = wrdataDF.rename(columns={'WRID':'Position'})
cleanwrdataDF['YDS'] = cleanwrdataDF['YDS'].str.replace(',', '').astype(float)
cleanwrdataDF['ROST'] = cleanwrdataDF['ROST'].str.rstrip('%').astype(float)
cleanwrdataDF.dropna(inplace=True)
cleanwrdataDF
```

```python
rbdataroute = "../Resources/FantasyPros_Fantasy_Football_Statistics_RB.csv"
rbdataDF = pd.read_csv(rbdataroute)
rbdataDF
```

```python
rbdataDF['RBID'] = "RB"
cleanrbdataDF = rbdataDF.rename(columns={'RBID':'Position'})
cleanrbdataDF['RushYDS'] = cleanrbdataDF['RushYDS'].str.replace(',', '').astype(float)
cleanrbdataDF['ROST'] = cleanrbdataDF['ROST'].str.rstrip('%').astype(float)
cleanrbdataDF.dropna(inplace=True)
cleanrbdataDF
```

```python
tedataroute = "../Resources/FantasyPros_Fantasy_Football_Statistics_TE.csv"
tedataDF = pd.read_csv(tedataroute)
tedataDF
```

```python
tedataDF['TEID'] = "TE"
cleantedataDF = tedataDF.rename(columns={'TEID':'Position'})
cleantedataDF['CatchYDS'] = cleantedataDF['CatchYDS'].str.replace(',', '').astype(float)
cleantedataDF['ROST'] = cleantedataDF['ROST'].str.rstrip('%').astype(float)
cleantedataDF.dropna(inplace=True)
cleantedataDF
```

```python
cleanqbdataDF.to_csv("../cleanedResources/cleanqbStats.csv", index=False)
cleanwrdataDF.to_csv("../cleanedResources/cleanwrStats.csv", index=False)
cleanrbdataDF.to_csv("../cleanedResources/cleanrbStats.csv", index=False)
cleantedataDF.to_csv("../cleanedResources/cleanteStats.csv", index=False)
```

```python
import sqlite3
import pandas as pd
conn = sqlite3.connect('../SQLite/Fantasystats.sqlite')
# Read Excel files
df1 = pd.read_csv('../cleanedResources/cleanqbStats.csv')
df2 = pd.read_csv('../cleanedResources/cleanrbStats.csv')
df3 = pd.read_csv('../cleanedResources/cleanwrStats.csv')
df4 = pd.read_csv('../cleanedResources/cleanteStats.csv')
# Write to database
df1.to_sql('QBStats', conn, if_exists='replace', index=False)
df2.to_sql('RBStats', conn, if_exists='replace', index=False)
df3.to_sql('WRStats', conn, if_exists='replace', index=False)
df4.to_sql('TEStats', conn, if_exists='replace', index=False)
conn.close()
```
