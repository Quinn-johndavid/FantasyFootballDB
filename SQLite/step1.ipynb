```python
import pandas as pd 
```

## Step 1: Preclean

```python
qbdataroute = "../dirtyresources/FantasyPros_Fantasy_Football_Statistics_QB.csv"
qbdataDF = pd.read_csv(qbdataroute)
qbdataDF
```

```python
renamedqbdatadf = qbdataDF.rename(columns={'Rank':'QBID', 'ATT':'PassingATT', 'YDS':'PassingYDS', 'Y/A': 'YardsPerThrow', 'TD':'PassingTD', 'ATT.1':'RushingATT', 
                                            'YDS.1': 'RushingYDS', 'TD.1': "RushingTD", "FL": "FumblesLost", 'G': 'Gamesplayed', 'FPTS/G': 'FPTSPerGame'})
renamedqbdatadf
```

```python
wrdataroute = "../dirtyresources/FantasyPros_Fantasy_Football_Statistics_WR.csv"
wrdataDF = pd.read_csv(wrdataroute)
wrdataDF
```

```python
renamedwrdatadf = wrdataDF.rename(columns={'Rank':'WRID', 'YDS':'RecYDS', 'Y/R': 'YardsPerReception','LG': 'Longest', '20+': '20YardsorMore', 'TD':'RecTD', 'ATT':'RushingATT', 
                         'YDS.1': 'RushingYDS', 'TD.1': "RushingTD", "FL": "FumblesLost", 'G': 'Gamesplayed', 'FPTS/G': 'FPTSPerGame'})
renamedwrdatadf
```

```python
rbdataroute = "../dirtyresources/FantasyPros_Fantasy_Football_Statistics_RB.csv"
rbdataDF = pd.read_csv(rbdataroute)
rbdataDF
```

```python
renamedrbdatadf = rbdataDF.rename(columns={'Rank':'RBID','ATT':'RushingATT', 'YDS':'RushingYDS', 'Y/A': 'YardsPerATT','LG': 'Longest', '20+': '20YardsorMore', 'TD':'RushingTD', 
                                            'YDS.1': 'RecYDS', 'TD.1': "RecTD", "Y/R": "YardsPerRec", "FL": "FumblesLost", 'G': 'Gamesplayed', 'FPTS/G': 'FPTSPerGame'})
renamedrbdatadf
```

```python
tedataroute = "../dirtyresources/FantasyPros_Fantasy_Football_Statistics_TE.csv"
tedataDF = pd.read_csv(tedataroute)
tedataDF
```

```python
renamedtedatadf = tedataDF.rename(columns={'Rank':'TEID', 'YDS':'RecYDS', 'Y/R': 'YardsPerReception','LG': 'Longest', '20+': '20YardsorMore', 'TD':'RecTD', 'ATT':'RushingATT', 
                         'YDS.1': 'RushingYDS', 'TD.1': "RushingTD", "FL": "FumblesLost", 'G': 'Gamesplayed', 'FPTS/G': 'FPTSPerGame'})
renamedtedatadf
```

```python
kickdataroute = "../dirtyresources/FantasyPros_Fantasy_Football_Statistics_K.csv"
kickdataDF = pd.read_csv(kickdataroute)
kickdataDF
```

```python
renamedkickDataDF = kickdataDF.rename(columns={ 'Rank': 'KickID', 'FG': 'FG Made', 'FGA': 'FG Attempted', 'LG': 'Longest', 'XPT': 'Extra Point Made', 'XPA': 'Extra Points Attempted', 'G': 'Games Played', 'FPTSPerGame': 'FPTSPerGame' })
renamedkickDataDF
```

```python
renamedqbdatadf.to_csv("../Resources/FantasyPros_Fantasy_Football_Statistics_QB.csv", index=False)
renamedwrdatadf.to_csv("../Resources/FantasyPros_Fantasy_Football_Statistics_WR.csv", index=False)
renamedrbdatadf.to_csv("../Resources/FantasyPros_Fantasy_Football_Statistics_RB.csv", index=False)
renamedtedatadf.to_csv("../Resources/FantasyPros_Fantasy_Football_Statistics_TE.csv", index=False)
renamedkickDataDF.to_csv("../Resources/FantasyPros_Fantasy_Football_Statistics_K.csv", index = False)
```
